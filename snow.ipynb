```python
notes_text = analyst_ratings["NOTES"].dropna().astype(str).tolist()

TRANSCRIPT_COL = None
for c in transcripts.columns:
    if "text" in c.lower() or "content" in c.lower() or "body" in c.lower():
        TRANSCRIPT_COL = c
        break

print("Using transcript column:", TRANSCRIPT_COL)

transcript_text = []
if TRANSCRIPT_COL:
    transcript_text = transcripts[TRANSCRIPT_COL].dropna().astype(str).tolist()

corpus = notes_text + transcript_text
len(corpus)
```

```python
topic_keywords = {
    "ai": ["ai", "gen ai", "llm", "cortex", "model", "machine learning", "vector", "copilot"],
    "pricing": ["pricing", "discount", "pressure", "competition", "rates", "seat", "cost"],
    "macro": ["macro", "economy", "budgets", "optimization", "spending", "headwinds", "slowdown"],
    "competition": ["aws", "azure", "google", "oracle", "datadog", "mongodb", "share"],
    "demand_visibility": ["rpo", "backlog", "pipeline", "duration", "renewal", "commitment"],
    "profitability": ["margin", "free cash flow", "fcf", "profit", "efficiency"]
}
```

```python
import re
from collections import Counter, defaultdict

def clean(text):
    return re.sub(r"[^a-zA-Z0-9 ]", " ", text.lower())

scores = defaultdict(int)

for doc in corpus:
    text = clean(doc)
    for topic, words in topic_keywords.items():
        for w in words:
            scores[topic] += text.count(w)

topic_strength = pd.Series(scores).sort_values(ascending=False)
topic_strength
```

```python
top_topics = topic_strength.head(3)
top_topics
```

```python
snow_g = growth_view["Snowflake QoQ Growth (%)"].iloc[0]

peer_summary = ", ".join(
    growth_view.sort_values("Peer QoQ Growth (%)", ascending=False)
    .head(3)
    .apply(lambda r: f"{r['Company']} ({r['Peer QoQ Growth (%)']:.1f}%)", axis=1)
    .tolist()
)

snow_rpo_b = snow_rpo_m / 1000

def generate_ai_questions(top_topics):
    qs = []

    qs.append(
        f"Peer performance: Snowflake grew {snow_g:.1f}% QoQ versus {peer_summary}. "
        "What operational or product drivers explain Snowflakes relative outperformance this quarter?"
    )

    for topic in top_topics.index:
        if topic == "ai":
            qs.append("AI adoption is a dominant theme in analyst commentary. How is Snowflake monetizing AI workloads and what early traction metrics matter most?")
        if topic == "macro":
            qs.append("Macro and cloud spend optimization remain in focus. What signals are you seeing in consumption trends and enterprise budgets?")
        if topic == "pricing":
            qs.append("Competitive pricing pressure is frequently mentioned. Are discounts increasing, and how is Snowflake protecting margins?")
        if topic == "demand_visibility":
            qs.append(f"RPO is roughly ${snow_rpo_b:.1f}B this quarter. How should investors think about backlog conversion and deal duration trends?")
        if topic == "competition":
            qs.append("Peers and hyperscalers are aggressively investing. Where is Snowflake gaining or losing share by workload type?")
        if topic == "profitability":
            qs.append("Profitability and free cash flow are key themes. What efficiency levers are driving margin improvements?")

    return list(dict.fromkeys(qs))[:5]

questions_aiish = generate_ai_questions(top_topics)
questions_aiish
```

```python
def ai_answer(q):
    base = (
        f"Snowflake delivered ${snow_rev_m:.1f}M revenue with {snow_g:.1f}% QoQ growth this quarter, "
        f"supported by RPO of about ${snow_rpo_b:.1f}B. "
        f"Top AI-extracted themes from unstructured data were: "
        + ", ".join(top_topics.index) + ". "
    )

    if "AI" in q or "ai" in q.lower():
        return base + "Management should emphasize expanding AI workloads, consumption leverage, and early adoption metrics."
    if "macro" in q.lower():
        return base + "Management should acknowledge optimization behavior while highlighting resilience in large enterprise demand."
    if "pricing" in q.lower():
        return base + "Management should frame pricing discipline and differentiation as margin protection tools."
    if "RPO" in q:
        return base + "Management should explain backlog conversion drivers and renewal stability."
    if "competition" in q.lower():
        return base + "Management should cite competitive wins and workload-level differentiation."
    return base + "Management should link performance drivers to repeatable growth levers."

qa_df = pd.DataFrame({
    "Question": questions_aiish,
    "AI-Generated Answer": [ai_answer(q) for q in questions_aiish]
})

qa_df
```
